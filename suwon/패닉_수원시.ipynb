## 10. 옥외광고물현황

```python
fig, ax = plt.subplots()
ax.boxplot([옥외광고물현황['가로'], 옥외광고물현황['세로'], 옥외광고물현황['높이']], sym="bo")
plt.title('Box plot for Outlier')
plt.xticks([1, 2, 3], ['width', 'length', 'height'])

plt.show()
```

```python
from collections import Counter 

def detect_outliers(df, q1, q3, features): 
    outlier_indices = [] 
    
    for col in features: 
        Q1 = np.percentile(df[col], q1)   # 삭제되는 거 너무 많아서 바꿈...
        Q3 = np.percentile(df[col], q3) 
        IQR = Q3 - Q1 
        
        outlier_step = 1.5 * IQR
        
        outlier_list_col = df[(df[col] < Q1 - outlier_step) | (df[col] > Q3 + outlier_step)].index 
        outlier_indices.extend(outlier_list_col) 
    
    outlier_indices = Counter(outlier_indices) 
    multiple_outliers = list(k for k, v in outlier_indices.items())
    
    return multiple_outliers 

Outliers_to_drop = detect_outliers(옥외광고물현황, 15, 85, ['가로','세로','높이'])
print(len(Outliers_to_drop))
```

```python
옥외광고물현황 = 옥외광고물현황.drop(Outliers_to_drop, axis = 0).reset_index(drop=True)
```

```python
fig, ax = plt.subplots()
ax.boxplot([옥외광고물현황['가로'], 옥외광고물현황['세로'], 옥외광고물현황['높이']], sym="bo")
plt.title('Box plot for Outlier')
plt.xticks([1, 2, 3], ['width', 'length', 'height'])

plt.show()
```

- 우선 Q1=15%, Q3=85% 로 IQR 이외 값 삭제했는데 기준 정해서 그 값 이상은 삭제하는 거로 해야할 듯
